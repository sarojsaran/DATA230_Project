```python
df_gdp = pd.read_csv('GDP Countries 1960-2020.csv')
```

```python
df_gdp.shape[0]
```

```python
for i in range(0,271116):
    team = df.iloc[i]['Team'];
    year = df.iloc[i]['Year'];
    print("processing row:", i)
    clear_output(wait=True)
    if (year < 1960 or year > 2020):
        continue
    for j in range(0,266):
        if (df_gdp.iloc[j]['Country Name'] == team):
            #print(team, year, df_gdp.iloc[j][str(year)])
            df['GDP'][i] = df_gdp.iloc[j][str(year)]
            break
```

```python
df.to_csv('new_combined_gdp.csv')
```
